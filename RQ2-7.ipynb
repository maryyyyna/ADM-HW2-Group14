1. What is the most common time in which users publish their posts?

```python
posts = pd.read_csv('instagram_posts.csv', delimiter='\t', nrows=100000)
posts_cts = posts['cts']
posts_cts = posts_cts.dropna()
```

```python
posts_cts.head()
```

```python
posts_cts = pd.to_datetime(posts_cts)
```

```python
posts_cts['time'] = posts_cts.dt.strftime('%H:%M:%S')
```

```python
dic={}
for t in posts_cts['time']:
    if t in dic:
        dic[t]+=1
    else:
        dic[t]=1
```

```python
most_common_time = sorted(dic.items(), key=lambda item: item[1], reverse = True)
#What is the most common time in which users publish their post
print('The most common time in which users publish their posts is ' + str(most_common_time[0][0]))
```

2. Create a function that receives a time intervals list as a parameter and returns a plot with the number of posts for each given interval.

```python
def n_posts_for_interval(lis):
    n=len(lis)
    labels=[]
    counts=np.zeros(n)
    for i in range(n):
        labels.append(lis[i][0]+'-'+lis[i][1])
        for t in posts_cts['time']:
            start, end=lis[i]
            if start <= t <= end:
                counts[i]+=1
      
    plt.figure(figsize=(15,5))
    x = range(n)
    plt.xticks(x, labels, rotation=50)
    plt.bar(x, counts, width = 0.4)
    plt.xlabel('Time intervals')
    plt.ylabel('Number of posts')
    plt.title('Number of posts for each interval')
    plt.show()
    return 
```

3. Create a function that receives a time intervals list as a parameter and returns a plot with the number of posts for each given interval.

```python
time_intervals = [('06:00:00','10:59:59'),('11:00:00','13:59:59'), ('14:00:00','16:59:59'), ('17:00:00','19:59:59'),
         ('20:00:00', '23:59:59'),('00:00:00', '02:59:59'), ('03:00:00','05:59:59')]
n_posts_for_interval(time_intervals)
```

1. What's the probability that a post receives more than 20% "likes" of the number of followers a user has?

```python
profiles = pd.read_csv('instagram_profiles.csv', delimiter='\t',usecols=['followers','profile_id'])
profiles
```

```python
from tqdm import tqdm
```

```python
top_rows = []
for chunk in tqdm(pd.read_csv('instagram_posts.csv',usecols=['numbr_likes','profile_id'], delimiter='\t',iterator=True, chunksize=500000)):
    top_rows.append(chunk)
```

```python
p=[]
for i in range(0,len(top_rows)):
    merged=pd.merge(top_rows[i],profiles,on='profile_id')
    prob=sum(merged['numbr_likes']>=merged['followers']*0.2)/len(merged)
    p.append(prob)
```

```python
sum(p)/len(p)
```

The probability that a post receives more than 20% "likes" of the number of followers a user has it's ~5.9%.

2. Do users usually return to locations? Extract the probability that a user returns to a site after having posted it in the past. Does that probability make sense to you? Explain why or why not.

```python
top_rows = []
for chunk in tqdm(pd.read_csv('instagram_posts.csv',usecols=['location_id','profile_id'], delimiter='\t',iterator=True, chunksize=500000)):
    top_rows.append(chunk)
```

```python
p=[]
for i in range(0,len(top_rows)):
    dups=sum(top_rows[i].pivot_table(index=['profile_id','location_id'],aggfunc='size')>1)
    prob=dups/len(top_rows[i].pivot_table(index=['profile_id','location_id'],aggfunc='size'))
    p.append(prob)
```

```python
sum(p)/len(p)
```

Usually users doesn't return to locations, it happens with ~ 15% probability and doesn't really make sense. A user can get back to a place he/she visited and not necessarily post again the location.
